```python
import matplotlib.pyplot as plt
import numpy as np

from spiketrain.kernel.fun import KernelFun
from spiketrain.sptr import SpikeTrains
```

## Compute ISI distribution

```python
# Generate samples from a homogeneous Poisson process
st = SpikeTrains.poisson_process_hom(rate=2e1, 
                                     time_end=1e0, 
                                     dt=1e-3, 
                                     ntrains=20)

# Compute all the interspike intervals
isis = st.interspike_intervals()

fig, (ax1, ax2) = plt.subplots(figsize=(6, 3), ncols=2)
st.plot(ax=ax1)
ax2.hist(isis)
```

## Compute PSTH

```python
# Generate samples from an inhomogeneous Poisson process
dt = 1e-3
t = np.arange(0, 1, dt)
tau = 0.15
rate = np.exp(6 * t / tau * np.exp(-t / tau) + 1.5)

st = SpikeTrains.poisson_process_inh(rate=rate, 
                                     dt=1e-3, 
                                     ntrains=20)

# Compute the psth using a Gaussian kernel
ker = KernelFun.gaussian_delta(delta=5e-3)
psth = st.psth(ker)

fig, (ax1, ax2) = plt.subplots(figsize=(6, 6), nrows=2)
st.plot(ax=ax1)
ax2.plot(t, rate, label='rate')
ax2.plot(t, psth, label='psth')
ax2.legend()
```

## Compute similarity measures

```python
# Generate samples from an inhomogeneous Poisson process
dt = 1e-3
t = np.arange(0, 1, dt)
tau = 0.15
rate = np.exp(6 * t / tau * np.exp(-t / tau) + 1.5)

st1 = SpikeTrains.poisson_process_inh(rate=rate, 
                                      dt=1e-3, 
                                      ntrains=20)
st2 = SpikeTrains.poisson_process_inh(rate=rate, 
                                      dt=1e-3, 
                                      ntrains=20)

ker = KernelFun.gaussian_delta(delta=5e-3)

# Compute distance-based similarity Md
md = st1.Md(st, ker, ker)

# Compute angle-based similarity Ma
ma = st1.Ma(st, ker, ker)

print('Md is {:.3f} and Ma is {:.3f}'.format(md, ma))
```
